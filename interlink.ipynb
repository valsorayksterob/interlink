```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/valsorayksterob/interlink/main/armenian_pubs.csv', sep=',')
df
```

Although there are some repetitions among lower\upper cases, **Calumet** is the most favorite pub to go with 27 votes:

```python
df2 = df.pivot_table(values='Age', index=['Fav_Pub'], aggfunc='count', fill_value = '-')
df2.sort_values('Age', ascending=False)
```


For females it is **Calumet** which is favorite, for males — **Station:**

```python
df3 = df.pivot_table(values='Age', index=['Gender', 'Fav_Pub'], aggfunc='count', fill_value = '-')
df3.sort_values('Age', ascending=False)
```

**18-years old** are the most frequent respondents:

```python
df4 = df.pivot_table(values='Gender', index=['Age'], aggfunc='count', fill_value = '-')
df4.sort_values('Gender', ascending=False)
```

30 000 Drams is what the most people responded as they can spend:

```python
df5 = df.pivot_table(values='Gender', index=['WTS'], aggfunc='count', fill_value = '-')
df5.sort_values('WTS', ascending=False)
```

I am willing to see the correlation between the primary importantance and the median income of the visitors. Obvious correlation between Pricing and the income value:

```python
df6 = df.pivot_table(values='Income', index=['Prim_Imp'], aggfunc='median', fill_value = '-')
df6.sort_values('Income', ascending=False)
```
